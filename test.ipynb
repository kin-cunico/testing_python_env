## Importing pkgs and setting up env

```python
import numpy as np 
import matplotlib.pyplot as plt
import os
import pandas as pd

pwd = os.getcwd() 
```

## importing data

```python
data = pd.read_csv(pwd + '/data_imm_brasil.csv', encoding='latin-1', sep=';')
data.head()
```

### splitting data by port of arrival

```python
port_of_entry = data['Porto de Chegada']
port_of_entry.head()

number_of_entries = port_of_entry.value_counts()
```

#### plotting bar chart with port of entry

```python
figure = plt.figure(figsize=(20, 10))
plt.plot()
plt.bar(number_of_entries.index, number_of_entries.values)
plt.xticks(rotation=90)
plt.xlabel('Porto de Chegada')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.ylabel('Número de Entradas')
```

### plotting destination

```python
port_of_destination = data['Destino']
port_of_destination.head()
```

### comparing number of entries vs port of destination

```python
# Create a figure with a single subplot
fig, ax = plt.subplots(figsize=(12, 6))

# Plot the bar chart for both port_of_destination and port_of_entry
bar_width = 0.4  # Adjust the width as needed
bar_positions_dest = range(len(port_of_destination.value_counts()))
bar_positions_entry = [pos + bar_width for pos in bar_positions_dest]

ax.bar(bar_positions_dest, port_of_destination.value_counts().values, width=bar_width, label='Porto de Destino')
ax.bar(bar_positions_entry, port_of_entry.value_counts().values, width=bar_width, label='Porto de Chegada')

ax.set_xticks([pos + bar_width / 2 for pos in bar_positions_dest])
ax.set_xticklabels(port_of_destination.value_counts().index, rotation=90)
ax.set_xlabel('Portos')
ax.set_ylabel('Número de Entradas')
ax.set_title('Número de Entradas por Porto')
ax.legend()

# Show the plot
plt.show()
```

### plotting by port of origin

```python
port_of_origin = data['Porto de Embarque']
port_of_origin

number_of_entries_by_origin = port_of_origin.value_counts()
number_of_entries_by_origin

figure = plt.figure(figsize=(20, 10))

color = ['blue', 'red', 'green', 'yellow', 'orange', 'purple', 'pink', 'brown', 'gray', 'black']

plt.plot()
plt.bar(number_of_entries_by_origin.index, number_of_entries_by_origin.values, color=color)
plt.xlabel('Porto de Origem')
plt.xticks(rotation=90)
plt.ylabel('Número de Entradas')
plt.title('Número de Entradas por Porto de Origem', fontsize=20)
```
